```python
from collections import defaultdict
import statistics
import mutual_expectancy as me
import time
# average me for all sets of length 12 in the odyssey

def average_me_for_passage_length(passage_length, corpus_object):
    start = time.time()
    corpus = corpus_object.full_corpus
    passages_me = []
    starter_set = False
    passages_created = 0
    for ind,row in corpus.iterrows():
        passages_created = len(passages_me)
        if passages_created%100==0:
            print("Processed "+  str(passages_created) + " passages. Time elapsed: " + str((time.time()-start)))
        if not starter_set:
            starter_set = True
            starter_ind = ind
        current_passage = row['string']
        #only do this if same text and book
        if (ind+(passage_length-1)) < (len(corpus)+starter_ind):
            if corpus.loc[[ind+(passage_length-1)]]['text'].values[0]==row['text']:
                if corpus.loc[[ind+(passage_length-1)]]['book'].values[0]==row['book']:
                    for i in range(1,passage_length):
                        current_passage += "/"+ corpus.loc[[ind+i]]['string'].values[0]
                    passages_me.append(me.arithmetic_passage_expectancy(current_passage, corpus_object.expectancies, split_on="/"))
    print("Built passage list, calculating.")
    print("Average ME for sets of length ", passage_length, ":")
    answer = statistics.mean(passages_me)
    print(answer)
    return answer

def get_all_avg_me_passage_lengths(corpus):
    avgs = defaultdict(float)
    for i in range (6, 13):
        print("Working on passages of length ", i)
        avgs[i] = average_me_for_passage_length(i, corpus)
    return avgs

#all = get_all_avg_me_passage_lengths(corpus)
four = average_me_for_passage_length(4, corpus)
        
```

```python
import helpers
import letters
importlib.reload(helpers)
importlib.reload(letters)
def count_long_positions_word(word):
    upper = 0
    lower = 0
    for i in range(len(word)-1):
        j=i+1
        if (word[i] not in letters.vowels) and (word[j] not in letters.vowels):
            upper += 1
            if not (word[i] in letters.mutes and word[j] in letters.liquids):
                lower += 1
        elif word[i] in letters.double_consonants:
            upper += 1
            lower += 1
        elif (j==len(word)-1) and word[j] in letters.double_consonants:
            upper += 1
            lower += 1
    return upper, lower

                
def count_long_vowels_word(word):
    long_vowels = 0
    for i in range(len(word)-1):
        if (word[i] in letters.long_vowels) or (word[i]+word[i+1] in letters.dipthongs):
            long_vowels += 1
    return long_vowels


def count_long_positions_corpus(corpus, upper_or_lower='upper'):
    longs = 0
    n_longs = 0
    lowercase_longs = 0
    n_lowercase = 0
    uppercase_longs = 0
    n_uppercase = 0
    for index,row in corpus.iterrows():
        for token in row['string'].split(" "):
            if upper_or_lower=='upper':
                longs_to_add = count_long_positions_word(token)[0]
            else:
                longs_to_add = count_long_positions_word(token)[1]
            longs += longs_to_add
            n_longs += 1
            if token[0].isupper():
                uppercase_longs += longs_to_add
                n_uppercase += 1
            else:
                lowercase_longs += longs_to_add
                n_lowercase += 1
    return [longs, n_longs, lowercase_longs, n_lowercase, uppercase_longs, n_uppercase]

def count_long_vowels_corpus(corpus):
    longs = 0
    n_longs = 0
    lowercase_longs = 0
    n_lowercase = 0
    uppercase_longs = 0
    n_uppercase = 0
    for index,row in corpus.iterrows():
        for token in row['string'].split(" "):
            longs_to_add = count_long_vowels_word(token)
            longs += longs_to_add
            n_longs += 1
            if token[0].isupper():
                uppercase_longs += longs_to_add
                n_uppercase += 1
            else:
                lowercase_longs += longs_to_add
                n_lowercase += 1
    return [longs, n_longs, lowercase_longs, n_lowercase, uppercase_longs, n_uppercase] 
    
def count_double_consonants_by_book(corpus):
    header = ['text', 'book', 'proper_nouns', 'double_consonants']
    return header
```

```python
#want ME of repeated line sets compared to other sets that don't include those lines
def line_set_ME(length_of_set):
    #dataframe of repeated line sets of target length
    target_sets = corpus.repeated_line_sets[corpus.repeated_line_sets['set_size']==length_of_set]
    #set of all lines that appear in repeated line sets for easy inclusion checks
    target_lines = []
    for index,row in target_sets.iterrows():
        target_lines.append(row['lines'].split('/'))
    #dataframe of all other sets of size length_of_set
    other_sets = []
    #iterate over every row in the corpus
    for ind1,row1 in corpus.me_by_line_df.iterrows():
        current_set = ""
        include = True
        if row1['string'] not in target_lines:
            current_set = row1['string']
            for next_line in range(1, length_of_set+1):
                if ind1 + next_line < 27789: #don't go over the end of the corpus
                    following_line = corpus.me_by_line_df.iloc[ind1 + next_line]['string']
                    if following_line not in target_lines:
                        current_set = current_set + "/" + following_line
                    else:
                        include = False #line appears in target_lines
                else:
                    include = False #line set is not long enough
        if include:
            other_sets.append([length_of_set, current_set, row1['text'], row1['book'], row1['line'], corpus
                              .arithmetic_passage_expectancy(current_set,split_on='/')])
    return other_sets
    
def calc_me_all_line_sets(to_file=False)
    all_sets = []
    header = ['set_size', 'line_set', 'text', 'book', 'line', 'passage_ME']
    for i in range(2, 13):
        all_sets = all_sets + line_set_ME(i)
    all_sets_df = pd.DataFrame(all_sets, columns=header)
    if to_file:
        all_sets_df.to_csv('data/all_line_sets_me.csv')
    return all_sets_df

    
            
    
                        
repeated_line_set_ME(12)
```

```python
#words that appear in this book that don't appear in any other book
iliad_books = defaultdict(set)
odyssey_books = defaultdict(set)

iliad_unique = defaultdict(list)
odyssey_unique = defaultdict(list)

rows = []

for index,row in corpus.corpus.iterrows():
    tokens = row.string.split()
    if row['text'] == 'iliad':
        for token in tokens:
            iliad_books[row.book].add(token)
    else:
        for token in tokens:
            odyssey_books[row.book].add(token)

for book in iliad_books.keys():
    for token in iliad_books[book]:
        found = False
        for other_books in iliad_books.keys():
            if book != other_books:
                if token in iliad_books[other_books]:
                    found = True
                    break
        for other_books in odyssey_books.keys():
            if token in odyssey_books[other_books]:
                found = True
                break
        if not found:
            iliad_unique[book].append(token)

for book in odyssey_books.keys():
    print("Odyssey ", str(book))
    for token in odyssey_books[book]:
        found = False
        for other_books in odyssey_books.keys():
            if book != other_books:
                if token in odyssey_books[other_books]:
                    found = True
                    break
        for other_books in iliad_books.keys():
            if token in iliad_books[other_books]:
                found = True
                break
        if not found:
            odyssey_unique[book].append(token)

for item in iliad_unique.keys():
    for word in iliad_unique[item]:
        rows.append(['iliad', item, word])
for item in odyssey_unique.keys():
    for word in odyssey_unique[item]:
        rows.append(['odyssey', item, word])
header = ['text', 'book', 'word']
df = pd.DataFrame(rows, columns=header)
df.to_csv('data/unique_words_by_book_iliad17_removed.csv')
df
```
